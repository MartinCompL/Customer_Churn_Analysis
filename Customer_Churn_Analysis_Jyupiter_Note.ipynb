What is the proportion of the customers that are still using the banking services compared to those that have left in the period covered in the dataset? Is there a significant difference in the proportion that the bank authority should be worried about?

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset into a pandas dataframe
df = pd.read_csv("Main_Sample.csv")

# Group the data by gender and exited status
grouped = df.groupby(["Gender", "Exited"]).size().unstack()

# Create a bar chart to visualize the number of male and female customers who have exited and stayed with the bank
labels = ["F - Stayed", "F - Exited", "M - Stayed", "M - Exited"]
data = [grouped.loc["Female", 0], grouped.loc["Female", 1], grouped.loc["Male", 0], grouped.loc["Male", 1]]
colors = ["pink", "purple", "blue", "lightblue"]
plt.bar(labels, data, color=colors)

# Set the chart title and axes labels
total_female = grouped.loc["Female"].sum()
total_male = grouped.loc["Male"].sum()
total_customers = total_female + total_male
plt.title(f"Number of Customers Who Have Exited and Stayed with the Bank by Gender\nTotal: {total_customers} (F: {total_female}, M: {total_male})")
plt.xlabel("Gender and Exited Status")
plt.ylabel("Number of Customers")

# Add text labels to the bars showing the exact number of customers who exited and did not exit the bank
for i, d in enumerate(data):
    plt.text(i, d, str(d), ha="center", va="bottom")

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset into a pandas dataframe
df = pd.read_csv("Main_Sample.csv")

# Calculate the total number of customers who have exited and stayed with the bank
num_exited = len(df[df["Exited"] == 1])
num_stayed = len(df[df["Exited"] == 0])
total_customers = num_exited + num_stayed

# Calculate the proportion of customers who have exited the bank
proportion_exited = num_exited / total_customers

# Calculate the proportion of customers who are still using the bank
proportion_stayed = num_stayed / total_customers

# Create a pie chart to visualize the proportions of customers who have exited and stayed with the bank
labels = ["Exited", "Stayed"]
proportions = [proportion_exited, proportion_stayed]
colors = ["red", "lightblue"]
plt.pie(proportions, labels=labels, colors=colors, autopct=lambda p: f"{int(round(p * total_customers / 100))}\n({p:.1f}%)")
plt.title(f"Proportions of Customers Who Have Exited and Stayed with the Bank\nTotal Customers: {total_customers}\nExited: {num_exited}\nStayed: {num_stayed}")

# Show the chart
plt.show()

# Check if there's a significant difference in the proportions
if proportion_exited > proportion_stayed:
    print(f"There are {num_exited} customers who have exited the bank, {num_stayed} customers who are still using the bank, and a significant difference in the proportions. The bank authority should be worried about the number of customers leaving the bank.")
else:
    print(f"There are {num_exited} customers who have exited the bank, {num_stayed} customers who are still using the bank, and no significant difference in the proportions. The bank authority should not be worried about the number of customers leaving the bank.")

```

What are the characteristics and statistics (in terms of gender, age groups, and tenure etc,) of the customers that are more likely to complain?

```python
# Filter for customers who have complained
complained_customers = df[df['Complain'] == 1]

# Create a bar plot of the number of complaints by age group
age_bins = [0, 20, 30, 40, 50, 60, 70, 80, 90, 100]
complained_customers['AgeGroup'] = pd.cut(complained_customers['Age'], bins=age_bins)
sns.countplot(data=complained_customers, x="AgeGroup")
plt.title("Number of Complaints by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")

# Label bars with precise count values
for i, v in enumerate(complained_customers.groupby('AgeGroup').size()):
    plt.text(i, v, str(v), ha='center')

plt.show()

# Create a bar plot of the number of complaints by credit card ownership
sns.countplot(data=complained_customers, x="HasCreditCard")
plt.title("Number of Complaints by Credit Card Ownership")
plt.xlabel("Has Credit Card")
plt.ylabel("Count")

# Label bars with precise count values
for i, v in enumerate(complained_customers.groupby('HasCreditCard').size()):
    plt.text(i, v, str(v), ha='center')

plt.show()

# Create a horizontal box plot of balance by gender for those who complained
sns.boxplot(data=complained_customers, x="Balance", y="Gender", orient="h")
plt.title("Balance Distribution by Gender for Complaining Customers")
plt.xlabel("Balance")
plt.ylabel("Gender")
plt.show()

# Calculate the percentage of complaining customers by gender
gender_complaint_rate = complained_customers.groupby('Gender').size() / df.groupby('Gender').size() * 100
print("Percentage of complaining customers by gender:\n", gender_complaint_rate)

# Calculate the average age of complaining customers
avg_age_complainers = complained_customers['Age'].mean()
print("Average age of complaining customers: {:.2f}".format(avg_age_complainers))

# Calculate the average tenure of complaining customers by gender
avg_tenure_complainers = complained_customers.groupby('Gender')['Tenure'].mean()
print("Average tenure of complaining customers by gender:\n", avg_tenure_complainers)
```

Is there a significant difference between the credit scores of all
the customers that have complained and those who have not in
the period covered in the dataset?


```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Read the dataset from a CSV file
df = pd.read_csv("Main_Sample.csv")

# Calculate the mean credit score of customers who complained and those who did not
complainers_mean_score = df.loc[df['Complain'] == 1, 'CreditScore'].mean()
non_complainers_mean_score = df.loc[df['Complain'] == 0, 'CreditScore'].mean()

# Print the mean credit scores
print("Mean credit score of complainers:", complainers_mean_score)
print("Mean credit score of non-complainers:", non_complainers_mean_score)

# Calculate the two-sample t-test
t, p = ttest_ind(df.loc[df['Complain'] == 1, 'CreditScore'], df.loc[df['Complain'] == 0, 'CreditScore'], equal_var=False)

# Print the p-value
print("P-value for two-sample t-test:", p)

# Plot the mean credit scores using a bar chart with labeled bars
plt.bar(['Complainers', 'Non-Complainers'], [complainers_mean_score, non_complainers_mean_score])
plt.text(0, complainers_mean_score, round(complainers_mean_score, 2), ha='center', va='bottom')
plt.text(1, non_complainers_mean_score, round(non_complainers_mean_score, 2), ha='center', va='bottom')
plt.text(0.5, max(complainers_mean_score, non_complainers_mean_score) * 1.01, f"P-value: {p:.3f}", ha='center', va='bottom')
plt.title('Mean Credit Scores of Complainers and Non-Complainers')
plt.xlabel('Customer Type')
plt.ylabel('Mean Credit Score')
plt.show()
```

Develop a model to predict whether a customer will complain or not
given the historical customer records in the Main Sample file.
Evaluate the developed model using appropriate metrics and report
its performances accordingly. Remember to also discuss the
implications of the performance (error or accuracy) of the model
with respect to the banking business.

Once your model is finalised, use it to predict whether the bank
customers included in New Sample file, will complain about the
banking services based on their profile information in the dataset.
Tabulate the predicated results in your report, in the order the
records are arranged in the file. This will be compared with the
expected outcomes for assessing the goodness of your model.


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Load data
df = pd.read_csv('Main_Sample.csv')

# Define the independent variables
X = df.drop(['CustomerId', 'Complain'], axis=1)

# Convert categorical variables to dummy variables
X = pd.get_dummies(X, drop_first=True)

# Define the dependent variable
y = df['Complain']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and fit the logistic regression model
lr_model = LogisticRegression(random_state=42)
lr_model.fit(X_train, y_train)

# Predict the test data
y_pred = lr_model.predict(X_test)

# Evaluate the model using confusion matrix and classification report
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
print('\n')
print('Classification Report:')
print(classification_report(y_test, y_pred))
print('\n')

# Evaluate the model using accuracy score
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Load data
df = pd.read_csv('Main_Sample.csv')

# Define the independent variables
X = df.drop(['CustomerId', 'Complain'], axis=1)

# Convert categorical variables to dummy variables
X = pd.get_dummies(X, drop_first=True)

# Define the dependent variable
y = df['Complain']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and fit the decision tree model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Predict the test data
y_pred = dt_model.predict(X_test)

# Evaluate the model using confusion matrix and classification report
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
print('\n')
print('Classification Report:')
print(classification_report(y_test, y_pred))
print('\n')

# Evaluate the model using accuracy score
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')


```
